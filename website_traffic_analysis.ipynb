```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("website_traffic_dataset.csv")
```

```python
df.head()
```

```python
df["date"]=pd.to_datetime(df["date"],format="%Y-%m-%d")
#converting to datetime variable
```

```python
traffic_by_source = df.groupby("source")["sessions"].sum()
bounce_by_source = df.groupby("source")["bounce_rate"].mean()
day_by_session = df.groupby("date")["sessions"].sum()
```

```python
plt.plot(day_by_session.index,day_by_session.values,marker="o")
plt.xlabel("Days")
plt.ylabel("Session")
plt.title("Session by Day")
plt.xticks(rotation=30,ha="right")
plt.grid(True)
plt.show()
```

```python
plt.bar(traffic_by_source.index,traffic_by_source.values)
plt.xlabel("Sources")
plt.ylabel("Sessions")
plt.title("Session by Source")
plt.show()
```

```python
plt.bar(bounce_by_source.index,bounce_by_source.values)
plt.xlabel("Sources")
plt.ylabel("Bounce Rate (%)")
plt.title("Bounce Rate by Sources")
plt.show()
```

```python
#The analysis provides clear insights into daily traffic trends and source performance. Organic traffic stands out as the most effective channel with strong engagement metrics. The visualizations are clear and actionable.
```
